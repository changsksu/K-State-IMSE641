# Data from Github (the same EV car dataset)

```python
import pandas as pd
# method 2
df2 = pd.read_csv('https://raw.githubusercontent.com/sidsriv/Introduction-to-Data-Science-in-python/master/cars.csv', sep=',', na_values=".")
# the foloowing statement only retrieves the first 5 obs
df2.head()
```

```python
# make cars a dataframe
cars=df2
cars.describe()
```

```python
cars[5:11]
```

```python
# retrieve the first three columns from the dataframe cars
cars[['YEAR', 'Make', 'Model']].head()
```

```python
# the following format does not work
# instead use iloc method
cars[[0]]
```

```python
# the followin command takes in all rows from the first three columns
cars.iloc[:, 0:3].head()
```

```python
cars2=cars.sort_values(by='YEAR', ascending=False)
cars2.head()
```

```python
cars.plot('Make', '(kW)')
```
